```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv("analytics-by-language.csv", header=1, names=["date", "Python", "R", "Java"], parse_dates=[0], index_col="date", na_values="<1")
df2 = pd.read_csv("machinelearning-by-language.csv", header=1, names=["date", "Python", "R", "Java"], parse_dates=[0], index_col="date", na_values="<1")
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 5), dpi=200)

df1["Python"].plot(ax=ax1)
df1["R"].plot(ax=ax1)
df1["Java"].plot(ax=ax1)
ax1.legend(["\"Python analytics\"", "\"R analytics\"", "\"Java analytics\""])
ax1.set_xlabel("")
ax1.set_xticks([])
ax1.set_ylim(0, 110)

df2["Python"].plot(ax=ax2)
df2["R"].plot(ax=ax2)
df2["Java"].plot(ax=ax2)
ax2.legend(["\"Python machine learning\"", "\"R machine learning\"", "\"Java machine learning\""])
ax2.set_ylim(0, 110)
ax2.set_xlabel("")

plt.subplots_adjust(hspace=0.05)
fig.text(0.07, 0.5, "Relative Google search volume", va="center", rotation="vertical")
```
